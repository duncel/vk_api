Разберем действия, которые будем выполнять.

**Ограничение:**
* за один запрос мы можем получить до 100 постов;
* максимальное смещение 100;
* запросы к ВК отправлять не чаще 3 запросов в секунду.

**Алгоритм:**
1. Смещение больше или равно количеству публикаций? Если нет, то п.2 \ Если да, то все посты обработаны
2. Отправляем запрос к ВК для получения последних 100 записей
3. Обрабатываем полученные данные (см. урок 4) и выводим на экран
4. Увеличиваем смещение на 100 и возвращаемся к 2 пункту

Код ниже позволяет получить информацию о количестве лайков, репостов, комментариев в любом сообществе или профиле. Данный пример не подойдет для групп и профилей с большим количеством постов. В таком случае лучше сразу писать ответы в файл.

````python
import requests
import time

access_token = 'f19c56cc59de71fe884aa6b08cf6680d9c75f2d323d5480cf76b1f0dc9d0aac21e8465ef46ccccbc8ec0f'
group_id = -62534510 #ID групп пишем с "-", профиль - без
count = 100 #количество постов, получаемых за 1 запрос
offset = 0 #начальное смещение

#Узнаем количество постов в группе

r = requests.get('https://api.vk.com/method/wall.get',
                 params={'owner_id':group_id,
                        'count':0,
                        'v':'5.95',
                        'access_token':access_token
                        })
response = r.json()
post_count = response['response']['count'] #количество постов в группе или профиле
print('Количество постов в сообществе:',post_count)

print('Дата публикации',',','Ссылка',',','Количество лайков',',','Количество репостов',',','Количество комментариев')

while offset < post_count:
    r = requests.get('https://api.vk.com/method/wall.get',
                 params={'owner_id':group_id,
                        'count':count,
                        'offset':offset,
                        'v':'5.95',
                        'access_token':access_token
                        })
    response = r.json()
    
    #запускаем цикл для обработкиответа
    for i in range(0,100):
        post_date = response['response']['items'][i]['date']
        comments_count = response['response']['items'][i]['comments']['count'] #количество комментариев
        post_id = response['response']['items'][i]['id'] #id поста в сообществе
        group_id = response['response']['items'][i]['from_id'] #id сообщества . В версиях API ниже 5.7 вместо поля owner_id приходит to_id.
        likes_count = response['response']['items'][i]['likes']['count'] #количество лайков
        repost_count = response['response']['items'][i]['reposts']['count']  #количество репостов
        print(post_date,',','https://vk.com/wall'+str(group_id)+'_'+str(post_id),',',likes_count,',',repost_count,',',comments_count)
        #print(post_date)
    offset = offset + 100
    time.sleep(.33) #задержка после первого запроса
````

````python
import requests
import datetime #импортируем пакеты
import time #импортируем пакеты

access_token = 'f19c56cc59de71fe884aa6b08cf6680d9c75f2d323d5480cf76b1f0dc9d0aac21e8465ef46ccccbc8ec0f'
group_id = -62534510 #ID групп пишем с "-", профиль - без
count = 100 #количество постов, получаемых за 1 запрос
offset = 0 #начальное смещение

#Узнаем количество постов в группе

r = requests.get('https://api.vk.com/method/wall.get',
                 params={'owner_id':group_id,
                        'count':0,
                        'v':'5.95',
                        'access_token':access_token
                        })
response = r.json()
post_count = response['response']['count'] #количество постов в группе или профиле
print('Количество постов в сообществе:',post_count)

print('Дата публикации',',','Ссылка',',','Количество лайков',',','Количество репостов',',','Количество комментариев')

while offset < post_count:
    r = requests.get('https://api.vk.com/method/wall.get',
                 params={'owner_id':group_id,
                        'count':count,
                        'offset':offset,
                        'v':'5.95',
                        'access_token':access_token
                        })
    response = r.json()
    
    #запускаем цикл для обработкиответа
    for i in range(0,100):
        post_date = datetime.datetime.fromtimestamp(int(response['response']['items'][i]['date'])).strftime('%Y-%m-%d') #форма преобразования Год-Месяц-День 
        post_date_h = datetime.datetime.fromtimestamp(int(response['response']['items'][i]['date'])).strftime('%H-%M-%S') #форма преобразования Час-Минута-Секунда 
        comments_count = response['response']['items'][i]['comments']['count'] #количество комментариев
        post_id = response['response']['items'][i]['id'] #id поста в сообществе
        group_id = response['response']['items'][i]['from_id'] #id сообщества . В версиях API ниже 5.7 вместо поля owner_id приходит to_id.
        likes_count = response['response']['items'][i]['likes']['count'] #количество лайков
        repost_count = response['response']['items'][i]['reposts']['count']  #количество репостов
        print(post_date,',',post_date_h,',','https://vk.com/wall'+str(group_id)+'_'+str(post_id),',',likes_count,',',repost_count,',',comments_count)
        #print(post_date)
    offset = offset + 100
    time.sleep(.33) #задержка после первого запроса
````
